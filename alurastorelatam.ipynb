### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Sumamos el precio total de cada tienda
ingreso1 = tienda['Precio'].sum()
ingreso2 = tienda2['Precio'].sum()
ingreso3 = tienda3['Precio'].sum()
ingreso4 = tienda4['Precio'].sum()

print(f"Tienda 1: ${ingreso1:,.2f}")
print(f"Tienda 2: ${ingreso2:,.2f}")
print(f"Tienda 3: ${ingreso3:,.2f}")
print(f"Tienda 4: ${ingreso4:,.2f}")

# Gráfico de barras para comparar
import matplotlib.pyplot as plt
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingreso1, ingreso2, ingreso3, ingreso4]

plt.bar(tiendas, ingresos, color=['blue', 'green', 'orange', 'red'])
plt.title('Ingresos Totales por Tienda')
plt.ylabel('Pesos ($)')
plt.show()
```

# 2. Ventas por categoría

```python
# Ejemplo para la Tienda 1 (puedes repetirlo para las demás)
print("Ventas por categoría en Tienda 1:")
print(tienda['Categoría del Producto'].value_counts())

# Gráfico circular de la Tienda 1
tienda['Categoría del Producto'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Categorías más vendidas - Tienda 1')
plt.ylabel('')
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Promedio de calificación por tienda
print(f"Calificación Tienda 1: {tienda['Calificación'].mean():.2f}")
print(f"Calificación Tienda 2: {tienda2['Calificación'].mean():.2f}")
print(f"Calificación Tienda 3: {tienda3['Calificación'].mean():.2f}")
print(f"Calificación Tienda 4: {tienda4['Calificación'].mean():.2f}")
```

# 4. Productos más y menos vendidos

```python
# Lo más vendido en Tienda 1
top_tienda1 = tienda['Producto'].value_counts().idxmax()
# Lo menos vendido en Tienda 1
bottom_tienda1 = tienda['Producto'].value_counts().idxmin()

print(f"En Tienda 1 el más vendido es: {top_tienda1}")
print(f"En Tienda 1 el menos vendido es: {bottom_tienda1}")
```

# 5. Envío promedio por tienda

```python
envio1 = tienda['Costo de envío'].mean()
envio2 = tienda2['Costo de envío'].mean()
envio3 = tienda3['Costo de envío'].mean()
envio4 = tienda4['Costo de envío'].mean()

print(f"Envío promedio Tienda 1: ${envio1:.2f}")
print(f"Envío promedio Tienda 4: ${envio4:.2f}")
```

**Visualización**

```python
import matplotlib.pyplot as plt

# Datos obtenidos del análisis
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [1150880300, 1116343500, 1098020000, 1038376400]
calificaciones = [3.98, 4.04, 4.05, 4.00]
envios = [26018, 25216, 24805, 23459]

# Configuración de estilo
plt.style.use('seaborn-v0_8-muted')

# Gráfico 1: Ingresos Totales (Barras)
plt.figure(figsize=(8, 5))
plt.bar(tiendas, ingresos, color='skyblue')
plt.title('Ingresos Totales por Tienda')
plt.ylabel('Pesos ($)')
plt.show()

# Gráfico 2: Calificación de Clientes (Línea)
plt.figure(figsize=(8, 5))
plt.plot(tiendas, calificaciones, marker='o', color='green', linewidth=2)
plt.title('Satisfacción del Cliente (Calificación Promedio)')
plt.ylim(3.8, 4.2)
plt.grid(True, alpha=0.3)
plt.show()

# Gráfico 3: Relación Ingresos vs Costo de Envío (Dispersión)
plt.figure(figsize=(8, 5))
plt.scatter(ingresos, envios, s=200, c=['blue', 'green', 'orange', 'red'])
for i, txt in enumerate(tiendas):
    plt.annotate(txt, (ingresos[i], envios[i]), xytext=(10,0), textcoords='offset points')
plt.title('Relación: Ingreso Total vs Costo de Envío')
plt.xlabel('Ingresos ($)')
plt.ylabel('Costo de Envío Promedio ($)')
plt.show()
```

Informe de Recomendación: Estrategia de Venta Alura Store

1. Introducción
El presente análisis tiene como objetivo evaluar el rendimiento de las cuatro sucursales de Alura Store para identificar la tienda menos eficiente. Se analizaron métricas financieras (ingresos), operativas (envíos) y de satisfacción (reseñas) para fundamentar la decisión de venta y capitalizar un nuevo emprendimiento para el Sr. Juan.

2. Desarrollo del Análisis
Tras procesar los datos de cada tienda, encontramos los siguientes hallazgos:

Ingresos: La Tienda 1 lidera en ventas ($1,150.8M), mientras que la Tienda 4 presenta la facturación más baja de la cadena ($1,038.3M).

Satisfacción: Las tiendas 2 y 3 tienen las mejores reseñas (4.04 y 4.05 respectivamente), lo que indica una clientela más fiel. La Tienda 1 tiene la calificación más baja (3.98).

Categorías y Productos: En todas las tiendas, la categoría de Muebles es la más popular. Sin embargo, productos como los "Instrumentos musicales" tienen muy poco movimiento en las Tiendas 3 y 4.

Logística: La Tienda 4 tiene el costo de envío promedio más bajo ($23,459), pero esto no compensa su bajo volumen de ingresos en comparación con las demás.

3. Conclusión y Recomendación
Mi recomendación técnica para el Sr. Juan es vender la Tienda 4.

Justificación:
Aunque la Tienda 1 tiene calificaciones ligeramente menores, es el motor de ingresos de la empresa. La Tienda 4, por el contrario, es la que menos dinero aporta al flujo total de la cadena. Vender esta sucursal permitirá obtener capital para el nuevo negocio con el menor impacto posible en los ingresos mensuales globales de Alura Store.

**Mapa de Dispersión Geográfica**

```python
# Asignar el nombre a cada DataFrame individual
tienda['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'
```

```python
import matplotlib.pyplot as plt

# Unimos todas las tiendas en un solo DataFrame (asegúrate de haber corrido las celdas anteriores)
df_total = pd.concat([tienda, tienda2, tienda3, tienda4])

# Creamos el mapa de puntos usando Longitud (X) y Latitud (Y)
plt.figure(figsize=(10, 8))
colores = {'Tienda 1': 'blue', 'Tienda 2': 'green', 'Tienda 3': 'orange', 'Tienda 4': 'red'}

for nombre_tienda, color in colores.items():
    subset = df_total[df_total['Tienda'] == nombre_tienda]
    plt.scatter(subset['lon'], subset['lat'], label=nombre_tienda, alpha=0.5, s=15, c=color)

plt.title('Distribución Geográfica de Ventas (Alura Store)')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
